# Stacked Contractive Autoencoder

```python
input_dim = x_train.shape[1]
input_dim
```

```python
import numpy as np
from keras.layers import Input, Dense, Reshape, Flatten, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras.optimizers import Adam
from keras.regularizers import l2
import matplotlib.pyplot as plt 
from keras.layers import Input, Dense
from keras import regularizers
from keras.losses import mse
import keras.backend as K

# Create the autoencoder
def create_autoencoder(input_shape, latent_dim, stack_depth):
    # Create input layer
    input_layer = Input(shape=input_shape)
    
    # Create encoding layers
    encoded = input_layer
    for i in range(stack_depth):
        encoded = Dense(input_shape[0] * 2, activation='relu', kernel_regularizer=l2(1e-4))(encoded)
    
    # Create decoding layers
    decoded = encoded
    for i in range(stack_depth):
        decoded = Dense(input_shape[0] * 2, activation='relu', kernel_regularizer=l2(1e-4))(decoded)
    
    # Create output layer
    output_layer = Dense(np.prod(input_shape), activation='sigmoid')(decoded)
    output_layer = Reshape(input_shape)(output_layer)
    
    # Compile the model
    autoencoder = Model(input_layer, output_layer)
    autoencoder.compile(optimizer=Adam(learning_rate=0.0001), loss='mse')
    
    return autoencoder

# Load your data here
data = x_train

# Define the shape of your input data    
input_shape = (118,)

# Define the latent dimension and stack depth
latent_dim = 64
stack_depth = 3

# Create the autoencoder
autoencoder = create_autoencoder(input_shape, latent_dim, stack_depth)

# Train the autoencoder
autoencoder.fit(x_train,x_train, epochs=10, batch_size=32, shuffle=True)
```

```python
import tensorflow as tf
encoder = tf.keras.Model(inputs=autoencoder.input,outputs=autoencoder.get_layer('dense_3').output)
x_features = encoder.predict(x_train)
x_features
```

```python
x_test_features = encoder.predict(x_test)
x_features.shape
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Create SVM model
svm_model = SVC(kernel='linear',C=1)

# Train the model
svm_model.fit(x_features, y_train)
```

```python
# Make predictions
y_pred = svm_model.predict(x_test_features)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```
